```python
x = pd.read_csv("delivery_orders_march.csv")
```

```python
x["pick"] = pd.to_datetime(x["pick"],unit='s')
x["1st_deliver_attempt"] = pd.to_datetime(x["1st_deliver_attempt"],unit='s')
x["2nd_deliver_attempt"] = pd.to_datetime(x["2nd_deliver_attempt"],unit='s')
x.head()
```

```python
x.insert(4,"buy",x["buyeraddress"])
```

```python
x.insert(5,"sell",x["selleraddress"])
```

```python
x.insert(6,"res",0)
```

```python
x["buy"] = x["buy"].apply(lambda s: s.split(" ")[-1].lower())
```

```python
x["sell"] = x["sell"].apply(lambda s: s.split(" ")[-1].lower())
```

```python
d = {"manila":{"manila":3,"luzon":5,"visayas":7,"mindanao":7},
     "luzon":{"manila":5,"luzon":5,"visayas":7,"mindanao":7},
     "visayas":{"manila":7,"luzon":7,"visayas":7,"mindanao":7},
     "mindanao":{"manila":7,"luzon":7,"visayas":7,"mindanao":7}}
```

```python
z = x.iloc[:20].copy(deep = True)
```

```python
def go(pick,first,second,time_allow):
    diff = (first - pick).days
    if diff >= 7:
        diff -= 1
    else:
        if first.dayofweek < pick.dayofweek:
            diff -= 1
    if diff >= time_allow:
        return 1
    else:
        if str(second) != "NaT":
            diff = (second - first).days
            if diff >= 7:
                diff -= 1
            else:
                if second.dayofweek < first.dayofweek:
                    diff -= 1
            if diff >= time_allow:
                return 1
            else:
                return 0
        else:
            return 0
        
```

```python
x["res"] = x.apply(lambda row: go(row['pick'], 
                                row['1st_deliver_attempt'],
                                row["2nd_deliver_attempt"],
                                d[row["buy"]][row["sell"]]), axis=1)
```
